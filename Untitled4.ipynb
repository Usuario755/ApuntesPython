# Cuantas E y e hay en una oracion?


*   Pide al usuario que ingrese una oracion.
*   Cuenta cuantas 'E' y cuantas 'e' hay en la oracion.
*   Imprime el resultado.
    



```python
oracion = input("Mensaje del usuario: ")
contadore = 0

for oracion in oracion:
  if oracion == 'E' or oracion == "e":
    contadore =+ 1

print(f"La oracion {oracion} tiene ({contadore})E y ({contadore})e")
```

#Adivina el numero
* Define un numero secreto.
* Pide al usuario que ingrese un numero.
* Indica si el numero es mayor o menor que el numero secreto.
* Repite hasta que el usuario adivine el numero.

```python
import random
i=0
numero_entero = random.randint(1, 100)
while i == 0:

  respuesta = int(input("Introducir numero (entre 1 a 100): "))


  if respuesta == numero_entero:
    print("¡Felicidades has acertado!")
    i=1;
  elif respuesta < numero_entero:
    print("El numero es mayor")

  elif respuesta > numero_entero:
    print("El numero es menor")


```
